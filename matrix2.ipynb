Eigenvectors and Eigenvalues

Eigenvectors are special vectors that does not change direction of the original vector when a matrix transformation is applied to the vector

Eigenvalue the the scaler quatity by which the original vector the changed i.e. streched or shinked

$$
A \mathbf{v} = \lambda \mathbf{v}
$$


```python
A = np.array([[1, 2], [2, -1]])
v = np.array([1, 4])
```

```python
eigValue, eigVector = np.linalg.eig(A)
```

```python
# EigenVector
eigVector 
```

```python
# EigenValue
eigValue
```

```python
np.dot(A,eigVector[:,0])
```

```python
np.dot(eigValue[0], eigVector[:, 0])
```

```python
np.dot(A, eigVector[:, 1])
```

```python
eigValue[1] * eigVector[:, 1]
```

Eigenvectors and eigenvalues using PyTorch

```python
AT = torch.tensor([[1, 2], [2, -1.]])
```

```python
eigenvalues_torch, eigenvectors_torch = torch.linalg.eigh(AT)
```

```python
torch.matmul(AT, eigenvectors_torch[:, 0])
```

```python
eigenvalues_torch[0] * eigenvectors_torch[:, 0]
```

3D Matrix - Eigenvectors and eigenvalues using PyTorch

```python
X = torch.tensor([[1, 2, 3], [9, 7, 7], [5, 7, 1.]])
```

```python
lx, xv = torch.linalg.eig(X)
lx = lx.real
xv = xv.real
```

```python
print(torch.matmul(X, xv[:, 0]))
print(lx[0] * xv[:, 0])
```

```python
print(torch.matmul(X, xv[:, 1]))
print(lx[1] * xv[:, 1])
```
