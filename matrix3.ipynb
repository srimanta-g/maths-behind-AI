**Single value decomposition**
<br>
<br>
Similar to eigenvalue decomposition but where eigenvalue decomposition works in square matrix (n x n), SVD words in rectangle matrix (m x n).
<br>
Singular vector -> Eigenvector <br>
Singular value -> Eigenvalue <br>

$$
    A = UDV^T
$$

<br>

U -> Left singular matrix (m x m) <br>
D -> Singular value <br>
V -> Right singular matrix (n * n) <br>

```python
import numpy as np
import torch
```

```python
P = torch.tensor([[1, 2], [3, 4], [5, 6.]])
```

```python
U, D, V = torch.linalg.svd(P)
```

```python
print(U)
print(D)
print(V)
```

```python
PP = torch.linalg.pinv(P)
```

**Trace Operator**
<br>

Sum of all diagonal elements of the matrix<br>
Sum of all eigenvalues

```python
B = torch.tensor([[1, 2], [3, 4.]])
```

```python
torch.trace(B)
```

```python
l, v = torch.linalg.eig(B)
```

```python
torch.sum(l).real
```
